Odległóść punktów

```python
import numpy as np
np.__version__
```

```python
v1 = np.array([-1, 4])
v1
```

```python
np.linalg.norm(v1)
```

```python
v2 = np.array([-10, 10, 13])
np.linalg.norm(v2)
```

```python
p = np.array([3, 0])
q = np.array([0, 4])

np.linalg.norm(p - q)
```

```python
p = np.array([3, 0, 1])
q = np.array([0, 4, 2])
np.linalg.norm(p - q)
```

Rozwiązywanie układów równań

```python
A = np.array([[2, 4], [1, -1]])
print(A)
```

```python
B = np.array([[10], [-1]])
print(B)
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
X
```

Funkcje statystyczne

```python
price = np.array([[12.40, 12.80, 11.90, 12.60, 1000],
                  [12.50, 13.00, 11.70, 12.20, 2000],
                  [12.20, 13.40, 12.20, 13.20, 1500]])
print(price)
```

```python
price.sum()
```

```python
price.sum(axis=0) # suma kolumn poszczególnych
```

```python
price.sum(axis=1) # suma wierszy poszczególnych
```

```python
np.min(price)
```

```python
np.max(price, axis=1)
```

```python
np.median(price)
```

```python
np.mean(price, axis=0)
```

```python
np.std(price, axis=0)
```

```python
np.var(price, axis=0)
```
